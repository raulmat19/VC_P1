TAREA: Crea una imagen con la textura del tablero de ajedrez

```python
# Creamos una imagen de fondo negro
gris_img = np.zeros((ancho,alto,1), dtype = np.uint8) 

# El tablero de ajedrez requiere 8 casillas por fila, ese serÃ¡ nuestro tamaÃ±o
blocksize = int(ancho/8)

# Creamos un bucle donde se analizarÃ¡ cada casilla, y si la suma de la posiciÃ³n x e y es divisible entre 2, esa casilla deberÃ¡ ser blanca
for y in range(0, 8):
    for x in range (0, 8):
        if ((y+x)%2 == 0):
            gris_img[y*blocksize:(y+1)*blocksize, x*blocksize:(x+1)*blocksize, 0] = 255
           

#Visualiza con mapa de color gris
plt.imshow(gris_img, cmap='gray') #Te animo a probar a eliminar la especificaciÃ³n del mapa de color
plt.show()
```

Evento de ratÃ³n para mostrar el valor RGB del pÃ­xel

```python
# FunciÃ³n que trata eventos de ratÃ³n 
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero     
    if event==cv2.EVENT_MOUSEMOVE:
        #Valores del pÃ­xel
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]

        px = x
        py = y
```

```python
vid = cv2.VideoCapture(0)

# TipografÃ­a
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa el manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
                 
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

TAREA:
Pintar cÃ­rculos en las posiciones del pÃ­xel mÃ¡s claro y oscuro de la imagen 

Para encontrar la posiciÃ³n del pÃ­xel mÃ¡s claro y oscuro, se puede recorrer la matriz de la imagen en busca de los valores que satisfaga esta condiciÃ³n. El problema es, que si bien funciona, para cada frame relentiza muchÃ­simo la ejecuciÃ³n. Empleando la documentaciÃ³n de OPENCV, se encuentra la funciÃ³n minMaxLoc(), el cual ya nos da las posiciones correspondientes y se pinta en cada frame un circulo en ellos, como una especie de tracking.

Primero, se muestra un ejemplo encontrando la posiciÃ³n del pÃ­xel mÃ¡s claro y oscuro de forma manual, y aplicado cada cierto nÃºmero de frames para no relentizar tanto la ejecuciÃ³n.
A continuaciÃ³n, se mostrarÃ¡ un ejemplo de la funciÃ³n minMaxLoc() y su uso correspondiente

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)

frame_cap = 0
brightest_pixel = ()
darkest_pixel = ()

while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        h, w, c = frame.shape

        if frame_cap % 25 == 0:
            max = 0
            min = 255
            for y in range(0, h, 5):
                for x in range(0, w, 5):
                    pixel = sum(frame[y, x, 0:3])
                    if pixel > max:
                        max = pixel
                        brightest_pixel = (x, y)
                    
                    elif pixel < min:
                        min = pixel
                        darkest_pixel = (x, y)

        frame_cap += 1
        cv2.circle(frame, brightest_pixel, 15, (0,0,0), 2)
        cv2.circle(frame, darkest_pixel, 15, (255,255,255), 2)

        # Muestra fotograma
        cv2.imshow('Tracking pixeles mas claro y oscuro', frame)
        # Detenemos pulsado ESC
        if cv2.waitKey(20) == 27:
            break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()
  
    if ret:  
        min, max, min_loc, max_loc = cv2.minMaxLoc(cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY))
        cv2.circle(frame, max_loc, 15, (0,0,0), 2)
        cv2.circle(frame, min_loc, 15, (255,255,255), 2)
        
        # Muestra fotograma
        cv2.imshow('Tracking pixeles mas claro y oscuro', frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Pop art
Salida de la webcam inspirada en Andy Warhol

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cÃ¡mara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resoluciÃ³n para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Separamos canales
        r = frame[:,:,0]
        g = frame[:,:,1]
        b = frame[:,:,2]

        #Jugamos
        tl[:,:,0] = r
        tl[:,:,1] = g
        tl[:,:,2] = b

        tr[:,:,0] = 255 - g
        tr[:,:,1] = r
        tr[:,:,2] = b
        
        bl[:,:,0] = b
        bl[:,:,1] = 255 - r
        bl[:,:,2] = g

        br[:,:,0] = r
        br[:,:,1] = g
        br[:,:,2] = 255 - b
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colecciÃ³n de cÃ­rculos

```python
vid = cv2.VideoCapture(0)

ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del cÃ­rculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma redimensionando a la mitad para que quepa en pantalla
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(cv2.CAP_DSHOW)

#Dimensiones de la cÃ¡mara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resoluciÃ³n para ocupar menos pantalla
w=int(w/2)
h=int(h/2)

vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*3,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
ml = collage[h:h+h,0:w]
mr = collage[h:h+h,w:w+w]
bl = collage[h+h:h+h+h,0:w]
br = collage[h+h:h+h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:

        r = frame[:,:,0]
        g = frame[:,:,1]
        b = frame[:,:,2]

        h2, w2, c2 = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w2/10), int(h2/10)), cv2.INTER_NEAREST)

        #Dimensiones reducidas
        h3, w3, c3 = down_frame.shape

        #Separamos canales
        r2 = down_frame[:,:,0]
        g2 = down_frame[:,:,1]
        b2 = down_frame[:,:,2]

        #Jugamos
        tl[:,:,0:3] = 0
        mr[:,:,0:3] = (19, 247, 224)
        bl[:,:,0:3] = 0

        for y in range(0,h3):
            for x in range(0,w3):
                # La suma de los valores RGB define el radio del cÃ­rculo
                rad = int ( (r2[y,x] + g2[y,x] + b2[y,x])/(10*5))
                cv2.circle(tl,(x*10 + 5, y*10 + 5), rad, (255, 255, 255), -1)
                cv2.circle(mr,(x*10 + 5, y*10 + 5), rad, (247, 19, 217), -1)
                cv2.circle(bl,(x*10 + 5, y*10 + 5), rad, (143, 14, 137), -1)

    
        tr[:,:,0] = int(np.random.randint(100, 255)) - g
        tr[:,:,1] = r
        tr[:,:,2] = b

        ml[:,:,0] = b
        ml[:,:,1] = g
        ml[:,:,2] = int(np.random.randint(100, 255)) - r

        br[:,:,0] = r
        br[:,:,1] = int(np.random.randint(100, 255)) - g
        br[:,:,2] = 255 - b
    
        # Muestra composicion
        cv2.imshow('Pop Art by Raul', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()

```
